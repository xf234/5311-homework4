```python
import pandas as pd
file_path = '/Users/xiaoyufan/Desktop/candy-data.csv'
df = pd.read_csv(file_path)
df
```

```python
def get_ingredients(row):
    ingredients = []
    if row['chocolate'] == 1:
        ingredients.append('chocolate')
    if row['fruity'] == 1:
        ingredients.append('fruity')
    if row['caramel'] == 1:
        ingredients.append('caramel')
    if row['peanutyalmondy'] == 1:
        ingredients.append('peanutyalmondy')
    if row['nougat'] == 1:
        ingredients.append('nougat')
    if row['crispedricewafer'] == 1:
        ingredients.append('crispedricewafer')
    if row['hard'] == 1:
        ingredients.append('hard')
    if row['bar'] == 1:
        ingredients.append('bar')
    if row['pluribus'] == 1:
        ingredients.append('pluribus')
    return ingredients

# Apply the function to each row of the DataFrame and create the new column
df['ingredients'] = df.apply(get_ingredients, axis=1)

# Display the DataFrame to verify the new column
df
```

```python
columns_to_drop = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus']
df.drop(columns=columns_to_drop, inplace=True)

# Display the DataFrame to verify the columns have been deleted
df.head()
```

```python
df.to_csv('/Users/xiaoyufan/Desktop/candy_data_new.csv', index=False)
```
